```python
def FFT(x):

    N = len(x) # N is the total number of samples
    array = [] # Creating a list to account for manupulations of each sample
    for i in range(N): # iterating over all the elements to implement FFT
      array.append(2.718281828459045**((-2j*3.14*i)/N))
    if N == 1:
        return x
    else:
        X_even = FFT(x[::2]) #Recursive function to implement FFT
        X_odd = FFT(x[1::2])
        factor = \
          array
      #Adding the positive and negative parts
        X = np.concatenate(
            [X_even+factor[:int(N/2)]*X_odd,
             X_even+factor[int(N/2):]*X_odd])
        return X
```

```python
import numpy as np #Importing Numpy library to create sinusoidal inputs
sr = 128 # sampling rate

ts = 1.0/sr # sampling interval

t = np.arange(0,1,ts)


x1 = 3*np.sin(2*np.pi*4*t) + np.sin(2*np.pi*7*t)

x2 = 8*np.sin(2*np.pi*3*t) + 3*np.sin(2*np.pi*9*t)

#Plotting input_1
plt.figure(figsize = (8, 6))
plt.plot(t, x1, 'r')
plt.ylabel('Amplitude')

plt.show()

#Plotting input_2
plt.figure(figsize = (8, 6))
plt.plot(t, x2, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
X=FFT(x1) #Applying FFT function to our input_1

# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')

# Get the one-sided specturm
n_oneside = N//2
# get the one side frequency
f_oneside = freq[:n_oneside]

# normalize the amplitude
X_oneside =X[:n_oneside]/n_oneside

plt.subplot(122)
plt.stem(f_oneside, abs(X_oneside), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('Normalized FFT Amplitude |X(freq)|')
plt.tight_layout()
plt.show()
```
